### Extracting year, month and day from date.
We will write three functions to extract year, month and day from date respectively and then apply these functions on each row.

```python
def extract_year(date_str):
    '''
    Given a string of format of "yyyy-MM-dd" extract year. Use split()
    If the input in null, return null. Code is already provided for this part. You just have to modify else part
    
    Don't forget to convert to integer using int() before returning.
    '''
    if pd.isnull(date_str):
        return np.NaN
    else:
        # Below line just returns 0 always. You have to change it to extract year from date_str using split.
        # Make sure to change to int type before returning. Tests will not pass otherwise.
        # Your implementation should not assume that year is always 4 digits. Some ill-formed dates could have less/more.
        year = date_str.split('-')[0]
        return int(year)

```

We will use three columns, `day`, `year` and `month` to create a new date column. If some `day`, `month`, `year` triple is not compatible, we will not be able to create a date object out of them and hence we will find "unclean"  data.  
Just like before we will write a function to create date given a triple, and call this method for each row using `apply`

```python
# Complete the function below. Make sure to return datetime object from this function.
def create_date(year, month, day):
    if pd.isnull(year) or pd.isnull(month) or pd.isnull(day):
        return np.nan
    else:
        year = int(year)
        month = int(month)
        day = int(day)
        # Change next line to create datetime from year, month and day.
        return datetime(year=year,month=month,day=day) # Change this line.
    
    
```

```python
# [20 points] This test will check if previous function is implemented correctly. This will run only after submission.
```

```python
# We are going to use the function you just wrote to create a new column.
# If the following line does not fail, then it means that all triples were valid dates otherwise there was some error.
menu["date_cleaned"] = menu.apply(lambda row: create_date(row["year"], row["month"], row["day"]), axis=1)
```
